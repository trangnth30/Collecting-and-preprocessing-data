```
import bs4
import requests
import pandas as pd
```

```
def get_page_content(url):
   page = requests.get(url,headers={"Accept-Language":"en-US"})
   return bs4.BeautifulSoup(page.text,"html.parser")
```

```
df = []
df = pd.DataFrame(df)
for i in range(1, 5, 1):
  print(i)
  url = f'https://vnexpress.net/giao-duc-p{i}' #link website
  #print(url)
  soup = get_page_content(url)
  #khởi tạo các list để chứa thông tin
  Title =[]
  Link = []
  #Trích xuất các thông tin cần thiết và thêm vào mảng tương ứng
  #s1 = soup.find('div', {'class': 'width_common list-news-subfolder has-border-right'})
  all_news = soup.findAll('h3', {'class': 'title-news'})
  for i in range(len(all_news)):
    news = all_news[i].find('a')
    title.append(news.get_text())
    link.append(news.get('href'))
  df_temp = pd.DataFrame({'title':title, 'link':link})
  df = pd.concat([df, df_temp], axis = 0, ignore_index=True)

df['label'] = 'Giáo dục'
```
